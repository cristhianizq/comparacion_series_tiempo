```python
Percentiles_Distrito['p60']=TablaMarClient[['MARCACIONES','Distrito']].reset_index().groupby('Distrito').quantile(0.60)
```

```python
TablaMarClient=TablaMarClient.merge(Percentiles_Distrito, left_on='Distrito', right_index=True, how='left')
TablaMarClient
```

```python
TablaMarClient['tipoChurn']=np.where(TablaMarClient['MARCACIONES']<TablaMarClient['p20'],'BUENO',np.where(TablaMarClient['MARCACIONES']<TablaMarClient['p60'],'REGULAR','MALO'))
TablaMarClient
```
